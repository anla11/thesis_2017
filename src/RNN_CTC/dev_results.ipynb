```python
from read_data import read_data, read_dict, read_listimg
from decoder import *

filename = '../dataset/train.txt'
idx_from = 4000
idx_to = -1

print ('Read test data')
image_list, title_list = read_data(filename = filename, limit = idx_to)
list_index = list(range(len(image_list)))[idx_from:]

char_dataset, lookup, decode_lookup = read_dict()
decoder = ArgMaxDecoder(char_dataset, decode_lookup)
```

```python
width_frame = [3,4,5,6]
model_name = ['rnn_ctc_ada_3_trial', 'rnn_ctc_ada_4_trial', 'rnn_ctc_ada_5_trial', 'rnn_ctc_ada_6_trial']
stop_epoch = [42, 60, 60, 60]
```

```python
for x in range(len(model_name)):
# for x in range(3, 4):
    wer_filename, cer_filename = 'wer_%d.txt' % width_frame[x], 'cer_%d.txt' % width_frame[x]
    cer_file = open(cer_filename, 'w')
    wer_file = open(wer_filename, 'w')
    print (model_name[x], wer_filename, cer_filename)
    
    for epoch in range(1, 1+stop_epoch[x]):
        path = '../model/%s/%s_%02d.pth' % (model_name[x], model_name[x], epoch)
        net = torch.load(path)
        wer, cer = test_model(net, image_list, title_list, list_index, decoder, width_frame[x], logfile = None)
#         wer, cer = 0.0, 1.0
        wer_file.write('%f\n' % wer)
        cer_file.write('%f\n' % cer)
```
